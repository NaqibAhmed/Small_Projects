## Matplot example workflow

```python
# 1. import matplotlib and get it ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Preparing data
x = [1,2,3,4]
y = [100,200,300,100]

# 3. Setup plot
fig, ax = plt.subplots(figsize=[10,10])

# 4. Adding data to ax
ax.plot(x, y)

# 5. Customizing plot
ax.set(title = "Sample Plot",
      xlabel =  "X-axis",
      ylabel = "Y-axis")
```

### Bar plots
creating plots from dictonary

```python
Peanut_butter_prices = {"Almond Butter":10,
                        "Peanur Butter":8,
                        "Cashew Butter":12}
fig, ax = plt.subplots(figsize=[5,5])
ax.bar(Peanut_butter_prices.keys() , Peanut_butter_prices.values())
ax.set(title="Peanut Butter Shop",
          ylabel = "Prices ($)");
```

```python
# lets make a horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(Peanut_butter_prices.keys()), list(Peanut_butter_prices.values()));
```

### Histograms

```python
# make data and plot a histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Subplots
Two options for subplots

```python
# option one
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=[10,5])
ax1.plot(x, x/2);
ax2.scatter(np.random.randint(5, size=5) , np.random.randint(5, size=5));
ax3.bar(Peanut_butter_prices.keys() , Peanut_butter_prices.values());
ax4.hist(np.random.randn(5000));
```

```python
# Option 2 
# using list indexing
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=[10,5])
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.randint(5, size=5) , np.random.randint(5, size=5));
ax[1,0].bar(Peanut_butter_prices.keys() , Peanut_butter_prices.values());
ax[1,1].hist(np.random.randn(5000));
```

## Plot using Pandas DataFrame

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Date Range"] = pd.date_range("1/1/2022" , periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x="Date Range", y="Total Sales");
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.columns
```

```python
# Scatter plot
car_sales.plot(x="Odometer (KM)" , y="Price" , kind="scatter");
```

```python
car_sales
```

```python
car_sales.plot(x="Make" , y="Price" , kind="bar");
```

```python
x = pd.Series(car_sales.groupby("Make").sum().index)
y = car_sales.groupby("Make").sum()["Price"]
fig , ax = plt.subplots()
ax.bar(x,y)
ax.set(ylabel = "Total price of each model car combined");
```

## Using OOPs method (fig, ax) method

```python
heart_disease = pd.read_csv(r"https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
heart_disease.to_csv("Heart_disease.csv")
heart_disease.head()
```

```python
over50 = heart_disease[heart_disease["age"]>50]
over50.head()
```

```python
# OOP method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
# OOP method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over50["age"],
                     y=over50["chol"],
                     c=over50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol level",
       xlabel="Age",
       ylabel="Cholesterol");

# adding legends
ax.legend(*scatter.legend_elements(), title="Target")

# add a horizontal line
ax.axhline(over50["chol"].mean(),
           linestyle = "--");
```
