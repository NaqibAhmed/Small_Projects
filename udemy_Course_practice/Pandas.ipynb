```python
#loc and iloc
# loc reffers to index
# iloc refers to position
```

```python
animals = pd.Series(["Python" , "Panda" , "Anaconda" , "Lion" , "Panther"],
                   index=[2,4,8,6,5])
animals
```

```python
animals.loc[4]
#returns pandas because it has index 4
```

```python
animals.iloc[4]
# returns panther bcz it has position 4
```

```python
animals.iloc[:3]
# returns location upto 3
```
